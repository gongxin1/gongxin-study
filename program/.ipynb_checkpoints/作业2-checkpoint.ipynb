### 1.1 问题一：将二进制的浮点数转化为十进制的数

首先将输入的浮点数转化为字符串，然后以'.'为分割线将它分为整数、小数两部分，最后只将小数部分进行计算
### 1.2程序

```python
def f(x):                                                #x为输入的浮点数
    string_a=str(x)                                      #str将浮点数x转化为字符串
    
    
    string_integer, string_decimal = string_a.split('.') #分离整数部分和小数部分
    
    sum=0
    for i in range(len(string_decimal)):
        sum = sum + 2**(-i-1)*(int(string_decimal[i]))
        if x>0:                                 #判断输入数是正还是负，赋值给sum
            sum=sum
        else:
            sum=-sum
    return sum
f(-0.10101*2**3)
```

### 2.1问题二：画图
### 2.2程序

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1.92, 2.08, num=80)

def poly(x):
    return (x-2)**9


def shang(x):
    a = np.array([1, -18, 144, -672, 2016,-4032, 5376, -4608, 2304, -512],dtype=np.float)
    n = len(x)
    p = np.ones(n)*a[0]
    for i in range(1, 10):
        p = x*p + a[i]*(1+np.sign(a[i])*2**(-52))   #给每个系数一个小扰动
    return p

def xia(x):
    a= np.array([1, -18, 144, -672, 2016,-4032, 5376, -4608, 2304, -512],dtype=np.float)
    n = len(x)
    p2 = np.ones(n)*a[0]
    for i in range(1, 10):
        p2 = x*p2 + a[i]*(1-np.sign(a[i])*2**(-52))  
    return p2

y2 = poly(x)
    
y3 = shang(x) 

y4 = xia(x)

plt.plot(x,y2,label="$(x-2)^9$")
plt.plot(x,y3,'.',label='upper')
plt.plot(x,y4,'.',label="lower")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1.85, 2.15, num=80)

def poly(x):
    return (x-2)**9


a= np.array([1, -18, 144, -672, 2016,-4032, 5376, -4608, 2304, -512],dtype=np.float)
n = len(x)
p = np.ones(n)*a[0]
bp = np.ones(n)*a[0]

for i in range(1, 10):
    p = x*p + a[i] 
    bp = np.abs(x)*bp + np.abs(a[i])
bp = 2*9*2**(-52)*bp 
sup = p + bp
inf = p - bp

    
y2 = poly(x)
    
plt.ylim(-1e-8, 1e-8)
plt.plot(x,y2,label="$(x-2)^9$")
plt.plot(x,sup,'.',label='upper')
plt.plot(x,inf,'.',label="lower")
plt.legend()
plt.show()
```
